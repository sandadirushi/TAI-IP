##Data visualization##

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'user_id': [1, 2, 3, 4, 5],
    'item_id': ['A', 'B', 'C', 'D', 'E'],
    'rating': [4, 5, 3, 2, 4]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10, 6))
plt.bar(df['item_id'], df['rating'], color='skyblue')
plt.xlabel('Item ID')
plt.ylabel('Rating')
plt.title('Rating by Item ID')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'user_id': [1, 2, 3, 4, 5],
    'item_id': ['A', 'B', 'C', 'D', 'E'],
    'rating': [4, 5, 3, 2, 4]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate total rating count
total_ratings = df['rating'].sum()

# Calculate rating percentage for each item
df['rating_percentage'] = df['rating'] / total_ratings * 100

# Plot
plt.figure(figsize=(8, 8))
plt.pie(df['rating_percentage'], labels=df['item_id'], autopct='%1.1f%%', startangle=140)
plt.title('Rating Distribution by Item ID')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'user_id': [1, 2, 3, 4, 5],
    'item_id': ['A', 'B', 'C', 'D', 'E'],
    'rating': [4, 5, 3, 2, 4]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot histogram
plt.figure(figsize=(8, 6))
plt.hist(df['rating'], bins=5, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'user_id': [1, 2, 3, 4, 5],
    'item_id': ['A', 'B', 'C', 'D', 'E'],
    'rating': [4, 5, 3, 2, 4]
}

# Create DataFrame
df = pd.DataFrame(data)

# Pivot the DataFrame to get ratings as values
pivot_df = df.pivot(index='user_id', columns='item_id', values='rating')

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_df, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Rating Heatmap')
plt.xlabel('Item ID')
plt.ylabel('User ID')
plt.show()
```
